```python
############# SECTION E #############
from IPython.display import display

order_item_merged = order_item_df.merge(items_df[['id', 'item_price']], left_on='item_id', right_on='id', how='left')

order_item_merged['total_price'] = order_item_merged['item_price'] * order_item_merged['quantity']

print("Average Total Price:")
display(order_item_merged['total_price'].mean())

print("\nMinimum Total Price:")
display(order_item_merged['total_price'].min())

print("\nMaximum Total Price:")
display(order_item_merged['total_price'].max())
```
